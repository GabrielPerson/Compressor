## Compressor

Implementação de um Compressor de imagens utilizando `python 3.6` e `OpenCV 3.4.2` para a disciplina *Introdução à Computação Visual* - UFMG 2018/2

## Iniciando

```python
import sys
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from scipy import fftpack
import matplotlib.image as mpimg

print("Python version: ", sys.version)
print("OpenCV version: ", cv.__version__)
```

## Lendo a Imagem

```python
img = cv.imread('band.jpg', 0)
print("Dimensions: ", img.shape)
print("Size: ", img.size)
```

## Imprimindo a imagem

```python
if (len(img.shape) == 3):
    plt.imshow(cv.cvtColor(img))
else:
    plt.imshow(img, cmap = 'gray')
plt.title("How Did We Get So Dark - Royal Blood")
plt.axis("off")
plt.xticks([])
plt.yticks([])
plt.show()    
```

## Inclusão de funções auxiliares

```python
# funções auxiliares
# Transformada do COS 2D
def get_2D_dct(img):
    return fftpack.dct(fftpack.dct(img.T, norm='ortho').T, norm='ortho')

# Inversa da transformada do COS 2D
def get_2d_idct(coefficients):
    return fftpack.idct(fftpack.idct(coefficients.T, norm='ortho').T, norm='ortho')

# Cálculo do RMSE
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())

# Cálculo do PSNR
def psnr(predictions, targets):
    rmsev = rmse(predictions, targets)
    return 20 * np.log10(255/rmsev)
```

## Cálculo da DCT e da IDCT

```python
img = img.astype(float)
img = img-128
step = 8

size = img.shape
IDCT = np.zeros(size, dtype=np.float)
DCT = np.zeros(size, dtype=np.float)

for y in range(0,size[0],step):
    for x in range(0,size[1],step):
        dct = get_2D_dct(img[y:y+step, x:x+step])
        idct = get_2d_idct(dct)
        IDCT[y:y+step, x:x+step] = idct
        DCT[y:y+step, x:x+step] = dct
    
img = img+128  
IDCT = IDCT+128

plt.imshow(img, cmap = 'gray', clim=(0, 255))
plt.title('Imagem original'), plt.xticks([]), plt.yticks([])
plt.show()

plt.imshow(DCT, cmap = 'gray', clim=(0, 255))
plt.title('DCT'), plt.xticks([]), plt.yticks([])
plt.show()

plt.imshow(IDCT, cmap = 'gray', clim=(0, 255))
plt.title('DCT Inversa'), plt.xticks([]), plt.yticks([])
plt.show()

# Salva imagem DCT
mpimg.imsave("dct.png", DCT, 0, 255, cmap='gray')

# Salva imagem DCT Inversa
mpimg.imsave("idct.png", IDCT, 0, 255, cmap='gray')

# Calculo do RMSE; Deve ser 0 pois a imagem ainda não foi comprimida
print ("RMSE: %5.3f" % rmse(img, IDCT))
```

## Matriz de compressão

Foi utilizada uma matriz de quantização comumente recomendada para compressões JPEG.

```python
np.set_printoptions(precision=3)
np.set_printoptions(suppress=True)

# exemplo arbitrário de matriz de quantização para compressão JPEG
QM = np.matrix([[16, 11, 10, 16, 24, 40, 51, 61], 
                [12, 12, 14, 19, 26, 58, 60, 55],
                [14, 13, 16, 24, 40, 57, 69, 56],
                [14, 17, 22, 29, 51, 87, 80, 62],
                [18, 22, 37, 56, 68, 109, 103, 77],
                [24, 35, 55, 64, 81, 104, 113, 92],
                [49, 64, 78, 87, 103, 121, 120, 101],
                [72, 92, 95, 98, 112, 100, 103, 99]])

QM = QM.astype(float)

useDC = True  # encode or not DC coefficient with DCT


print ("Matriz de quantização:")
print (QM)
```

## Compressão JPEG

```python
img = img-128
for y in range(0,size[0],step):
    for x in range(0,size[1],step):
        dct = get_2D_dct(img[y:y+step, x:x+step])
        odct = dct
        DC = dct[0,0]
        
        dct = np.divide(dct, QM)
        dct = dct.round()
        coddct = dct
        
        dct = np.multiply(dct, QM)
        if (useDC == True):
            dct[0,0] = DC         
        IDCT[y:y+step, x:x+step] = get_2d_idct(dct)
        
print ("\nPatch after DCT:"  )      
print (odct)

print ("\nPatch to be coded:")
print (coddct)

print ("\nReconstructed patch:")
print (dct ) 
```

## Imprime imagens

```python
IDCT = IDCT + 128  
img = img + 128
        
plt.imshow(img, cmap = 'gray', clim=(0, 255))
plt.title('Imagem original'), plt.xticks([]), plt.yticks([])
plt.show()

plt.imshow(DCT, cmap = 'gray', clim=(0, 255))
plt.title('DCT'), plt.xticks([]), plt.yticks([])
plt.show()

plt.imshow(IDCT, cmap = 'gray', clim=(0, 255))
plt.title('DCT inversa'), plt.xticks([]), plt.yticks([])
plt.show()

#file with decoded JPEG
mpimg.imsave("outJPEG.jpg", IDCT, 0, 255, cmap='gray')

print ("RMSE: %5.3f" % rmse(img, IDCT))
print ("PSNR: %5.3f" % psnr(img, IDCT))
```

```python
diff = img-IDCT
plt.imshow(diff, cmap = 'gray', clim=(0, 255))
plt.title('Subtração das imagens (original - comprimida)'), plt.xticks([]), plt.yticks([])
plt.show()
```
